# Scraping

## detik.com

```python
page = 1
text = []
tags = ['covid-19','pandemi','virus']


urllib3.disable_warnings()

while page <= 100:
    for tag in tags:
        url = f"https://www.detik.com/tag/{tag}/?sortby=time&page={page}"
        req = http.request("GET",url)
        soup = BeautifulSoup(req.data,'html.parser')
    
        #for i in soup.select('article h2'):
            #title.append(i.get_text(strip=True))
    
        for j in soup.select('article a span p'):
            text.append(j.get_text(strip=True))
    
    page += 1
```

```python
label = []
for i in range(len(text)):
    label.append('valid')
    
data = pd.DataFrame(text,columns=['text'])
data['label'] = label
```

## turnbackhoax

```python
page = 1
title = []
text = []
date = []
tag = ['covid-19','pandemi','virus']

urllib3.disable_warnings()

while page < 63:
    for tags in tag:
        url = f"https://turnbackhoax.id/page/{page}/?s={tags}"
        req = http.request("GET",url)
        soup = BeautifulSoup(req.data,'html.parser')
    
        for i in soup.select('h3 > a'):
            title.append(i.get_text(strip=True))
    
        for j in soup.select('article div p'):
            j = j.get_text()
            j = j.split()
            del(j[-1])
            text.append(" ".join(j).strip())
    
        for k in soup.find_all('span',class_='mh-meta-date updated'):
            date.append(k.get_text(strip=True))
        
    page += 1
```

```python
data = {'text':text,'label':label}
df = pd.DataFrame(data)
```

```python
df = df.drop_duplicates(subset=['text'],keep='first')
df = df.reset_index(drop=True)
```

```python
# export dataset ke excel
df.to_excel('E:/orbit_future_academy/data/turnbackhoax.xlsx',index = False)
```
